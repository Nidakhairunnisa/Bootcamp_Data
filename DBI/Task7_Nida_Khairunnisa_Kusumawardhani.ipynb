## **Tampilkan Barplot**

```python
usia = df.groupby(['umur'])['botak_prob'].count().sort_values(ascending=False).reset_index().head(10)
usia.rename(columns={'botak_prob':'Number of listings'},inplace=True)

plt.figure(figsize=(12, 6));
sns.barplot(usia['umur'],usia['Number of listings'],palette='rocket_r');
plt.title('Top 10 Peluang botak');
plt.xticks(rotation=70);
```

Dapat kita lihat bahwa peluang 10 besar umur yang berpeluang botak berada pada range 35-44 tahun

```python
sampo = df['sampo'].value_counts().reset_index()
sampo.rename(columns={'index':'sampo','sampo':'Number of Listings'},inplace=True)

plt.figure(figsize=(12, 6));
sns.barplot(sampo['sampo'],sampo['Number of Listings'],palette='crest');
plt.title('sampo vs Number of Listings');
```

Dari data survey yang kita dapat, dapat kita lihat bahwa Deadbuoy merupakan sampo yang paling banyak digunakan meskipun selisih perbandingan dengan merek sampo lainnya tidak begitu jauh.

```python
provinsi = df.groupby(['provinsi'])['botak_prob'].count().sort_values(ascending=False).reset_index().head(10)
provinsi.rename(columns={'botak_prob':'Number of listings'},inplace=True)

plt.figure(figsize=(12, 6));
sns.barplot(provinsi['provinsi'],provinsi['Number of listings'],palette='rocket_r');
plt.title('Top 10 Peluang botak');
plt.xticks(rotation=70);
```

Peluang Botak terbanyak berada di Provinsi Kupang

```python
pekerjaan = df.groupby(['pekerjaan'])['botak_prob'].count().sort_values(ascending=False).reset_index().head(10)
pekerjaan.rename(columns={'botak_prob':'Number of listings'},inplace=True)

plt.figure(figsize=(12, 6));
sns.barplot(pekerjaan['pekerjaan'],pekerjaan['Number of listings'],palette='rocket_r');
plt.title('Pekerjaan VS Number of listing');
plt.xticks(rotation=70);
```

Dapat kita lihat bahwa pekerjaan yang berpeluang botak terbanyak yaitu Pegawai Swasta

```python
jenis_kelamin = df.groupby(['jenis_kelamin'])['botak_prob'].count().sort_values(ascending=False).reset_index().head(10)
jenis_kelamin.rename(columns={'botak_prob':'Number of listings'},inplace=True)

plt.figure(figsize=(12, 6));
sns.barplot(jenis_kelamin['jenis_kelamin'],jenis_kelamin['Number of listings'],palette='rocket_r');
plt.title('jenis_kelamin VS botak_prob');
plt.xticks(rotation=70);
```

Laki-laki memiliki peluang terbesar untuk botak.

```python
pendidikan = df.groupby(['pendidikan'])['botak_prob'].count().sort_values(ascending=False).reset_index().head(10)
pendidikan.rename(columns={'botak_prob':'Number of listings'},inplace=True)

plt.figure(figsize=(12, 6));
sns.barplot(pendidikan['pendidikan'],pendidikan['Number of listings'],palette='rocket_r');
plt.title('Pendidikan VS peluang botak');
plt.xticks(rotation=70);
```

Pendidikan S1 memiliki peluang terbesar.

```python
stress = df.groupby(['stress'])['botak_prob'].count().sort_values(ascending=False).reset_index().head(10)
stress.rename(columns={'botak_prob':'Number of listings'},inplace=True)

plt.figure(figsize=(12, 6));
sns.barplot(stress['stress'],stress['Number of listings'],palette='rocket_r');
plt.title('stress VS peluang botak');
plt.xticks(rotation=70);
```

Stress memiliki pengaruh dalam peluang botak.

## **Dan handle seluruh masalah dengan tahap data preprocessing (data cleaning)**

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df.fillna({'umur': 0}, inplace=True)
df.fillna({'jenis_kelamin': 'tidak diketahui'}, inplace=True)
df.fillna({'provinsi': 'tidak diketahui'}, inplace=True)
df.fillna({'pekerjaan': 'tidak diketahui'}, inplace=True)
df.fillna({'gaji':0}, inplace=True)
df.fillna({'is_menikah':0}, inplace=True)
df.fillna({'is_keturunan':0}, inplace=True)
df.fillna({'sampo': 'tidak diketahui'}, inplace=True)
df.fillna({'pendidikan': 'tidak bersekolah'}, inplace=True)
df.fillna({'is_merokok': 0}, inplace=True)
df.fillna({'stress': 0}, inplace=True)
df.fillna({'botak_prob': 0}, inplace=True)
```

```python
df = df.drop('berat', axis=1)
df = df.drop('tinggi', axis=1)
```

```python
df.isnull().sum()
```

```python
df['umur'] = df['umur'].astype('int')
df['is_menikah'] = df['is_menikah'].astype('int')
df['is_keturunan'] = df['is_keturunan'].astype('int')
df['is_merokok'] = df['is_merokok'].astype('int')
df['stress'] = df['stress'].astype('int')
```
