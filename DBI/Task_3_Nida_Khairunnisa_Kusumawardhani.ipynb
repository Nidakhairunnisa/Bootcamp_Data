## 1. Buatlah 3 dataframe dengan dictionary, dimana dictionary 1 diberi nama PialaEropa dengan variable pertama adalah Negara dan isinya adalah jerman, spanyol, uni soviet, italia dan prancis lalu variable kedua yaitu piala dengan isinya adalah 3,3,1,1,2. Sedangkan dictionary 2 diberi nama PialaAsia dengan variable kedua adalah Negara dan isinya adalah jepang, indonesia, korea selatan, irak dan saudi arabia, lalu variable kedua yaitu piala dengan isinya 4, 0, 2, 1, 3. dan dictionary 3 diberi nama PialaDunia dengan variable pertama adalah Negara dan isinya adalah jerman, spanyol, korea selatan, jepang, dan prancis. lalu variable kedua yaitu piala dengan isinya adalah 4,1,0,0,3.

## Kemudian gabungkan ketiga dataframe tersebut dan diberi nama Pemenang. Lalu tampilkan berapa baris pada dataframe dan urutkan berdasarkan pemilik piala terbanyak pada dataframe tersebut !

```python
import pandas as pd
```

dictionary 1 diberi nama PialaEropa dengan variable pertama adalah Negara dan isinya adalah jerman, spanyol, uni soviet, italia dan prancis lalu variable kedua yaitu piala dengan isinya adalah 3,3,1,1,2.

```python
pialaEropa = pd.DataFrame({"negara" : ["jerman", "spanyol", "uni soviet", "italia", "prancis"], "piala" : [3,3,1,1,2]})
pialaEropa
```

dictionary 2 diberi nama PialaAsia dengan variable kedua adalah Negara dan isinya adalah jepang, indonesia, korea selatan, irak dan saudi arabia, lalu variable kedua yaitu piala dengan isinya 4, 0, 2, 1, 3.

```python
pialaAsia = pd.DataFrame({"negara" : ["jepang", "indonesia", "korea selatan", "irak", "saudi"], "piala" : [4,0,2,1,3]})
pialaAsia
```

dictionary 3 diberi nama PialaDunia dengan variable pertama adalah Negara dan isinya adalah jerman, spanyol, korea selatan, jepang, dan prancis. lalu variable kedua yaitu piala dengan isinya adalah 4,1,0,0,3.

```python
pialaDunia = pd.DataFrame({"negara" : ["jerman", "spanyol", "korea selatan", "jepang", "prancis"], "piala" : [4,1,0,0,3]})
pialaDunia
```

Kemudian gabungkan ketiga dataframe tersebut dan diberi nama Pemenang. Lalu tampilkan berapa baris pada dataframe dan urutkan berdasarkan pemilik piala terbanyak pada dataframe tersebut !

```python
pemenang = pd.concat([pialaEropa, pialaAsia, pialaDunia])
pemenang
```

```python
urutan_pemenang = pemenang.groupby(["negara"]).sum()[["piala"]]
```

```python
urutan_pemenang.sort_values("piala", ascending=False)
```

## 2. Lalu tampilkan index ke 0, 3, dan 5 dan simpan divariable baru dengan nama JuaraBertahan!

```python
juara_bertahan = urutan_pemenang.iloc[[0,3,5],:]
juara_bertahan
```

```python
piala_lebihdari_2 = urutan_pemenang[urutan_pemenang["piala"] >= 2]
piala_lebihdari_2.head()
```

## 5. Dengan menggunakan data banking pada tool script pada web Data Bangalore, dimana terdapat kasus pelanggan yang tidak memiliki rumah, belum menikah akan ditolak dalam pengajuan pinjaman. Pergunakan data yang ditolak pinjaman untuk dianalisis lebih lanjut dengan mengurutkan durasi pinjam paling lama dan sebutkan rata-rata umur peminjam nya?

```python
import pandas as pd
df5=pd.read_csv('/content/gdrive/MyDrive/Bootcamp Data Bangalore/banking.csv')
```

```python
no_housing=df5[df5['housing']=="no"]
no_housing
```

```python
single=df5[df5['marital']=="single"]
single
```

```python
pinjaman_tolak = pd.concat([no_housing, single])
pinjaman_tolak
```

```python
pinjaman_tolak.sort_values("duration", ascending=False)
```

```python
pinjaman_tolak.mean()
```

Dapat dilihat bahwa rata-rata usia pinjaman ditolak adalah 37 tahun.
